```python
df1 = df.dropna()
```

```python
df1 = df1.reset_index().drop("index", axis=1)
```

```python
df1.head()
```

```python
df1['budget'] = df1['budget'].astype('int')
```

```python
df1['gross'] = df1['gross'].astype('int')
df1['votes'] = df1['votes'].astype('int')
```

```python
pd.set_option('display.max_rows', None)
```

```python
df1
```

```python
neli = []
new_year = []
lis = []
for i in df1['released'].values:
    if type(i) == list:
        for z in i:
            lis.append(z)
    elif type(i) != list:
        lis.append(i)
for i in lis:
    y = i.split(' ')
    new_year.append(y)
for item in new_year:
    for i in item:
        if i.isdigit() and len(i) == 4:
            neli.append(i)
df1['year_corrected'] = neli
```

```python
df1.tail()
```

```python
df1 = df1.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df1[df1['budget'] < 0]
```

```python
gro = []
for i in df1['gross']:
    if i < 0:
        t=i*(-1)
        gro.append(t)
    else:
        gro.append(i)
df1['gross'] = gro
```

```python
df1.duplicated().sum()
```

##### Checking the correlation between Budget and Gross earning

```python


plt.scatter(x=df1['budget'], y=df1['gross'])
plt.title('budget vs gross earnings')
plt.ylabel('budget for film')
plt.xlabel('gross earnings')
plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df1, scatter_kws={"color": "red"}, line_kws={"color": "black"})
```

```python
correlation_matrix = df1.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('numeric features correlation matrix')
plt.ylabel('Movie Feature')
plt.xlabel('Movie Feature')

plt.show()
```

#### Checking for corrrelation between any other non-numeric values

```python
df2 = df1.copy()

for cols in df2.columns:
    if df2[cols].dtype == 'object':
        df2[cols] = df2[cols].astype('category')
        df2[cols] = df2[cols].cat.codes
df2
```

```python
correlation_matrix = df2.corr(method='pearson')

plt.figure(figsize = (15,8))

sns.heatmap(correlation_matrix, annot=True)

plt.title('numeric features correlation matrix')
plt.ylabel('Movie Feature')
plt.xlabel('Movie Feature')

plt.show()
```

```python
corr_mat = df2.corr()
cc = corr_mat.unstack()
cc
```
